```python
#Task 2
# Filter data for people with diabetes (Outcome = 1)
diabetes_patients = df[df['Outcome'] == 1]

# Calculate the median blood pressure
median_blood_pressure = diabetes_patients['BloodPressure'].median()
```

```python
#Task 3
# Identify the variable indicating diabetes diagnosis
diagnosis_variable = 'Outcome'  # The 'Outcome' column indicates diabetes diagnosis (0 or 1)

# Calculate the counts of each diagnosis category
diagnosis_counts = df[diagnosis_variable].value_counts()

# Calculate the proportion of each diagnosis category
diagnosis_proportions = diagnosis_counts / len(df)
print("\nDiagnosis Proportions:\n", diagnosis_proportions)
```

```python
#Task 4
X = df.drop('Outcome', axis=1)  # Features (all columns except 'Outcome')
y = df['Outcome']  # Target variable ('Outcome' column)

# Display shapes of X and Y
print("\nShape of X:", X.shape)
print("Shape of Y:", y.shape)
```

#Task 5
The train_test_split function is used to divide a dataset into training and testing sets.

test_size=0.2 → Allocates 20% of the data for testing and 80% for training.
random_state=1 → Ensures that the data split remains consistent every time the code is run.
stratify=y → Maintains the same class distribution in both training and testing sets, which is crucial for classification tasks, especially when dealing with imbalanced datasets.

```python
# Task 6
#modify the code from TASK5 so that a train-test split of 3:1 is obtained
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)

# Check the shape of the splits
print("Training set size:", X_train.shape, y_train.shape)
print("Testing set size:", X_test.shape, y_test.shape)

#The code splits the dataset into training and testing sets using a 3:1 ratio (75% for training, 25% for testing) while ensuring the target variable's distribution is maintained in both sets due to stratify=y. random_state is used for reproducibility
```

```python
# Task 7
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Create and Train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=1)  # Initialize the classifier
clf.fit(X_train, y_train)  # Train the classifier using the training data

# 2. Predict on the Training and Test Sets
y_train_pred = clf.predict(X_train)  # Predictions on the training data
y_test_pred = clf.predict(X_test)  # Predictions on the test data

# 3. Evaluate the Model's Performance
training_accuracy = accuracy_score(y_train, y_train_pred)  # Calculate training accuracy
testing_accuracy = accuracy_score(y_test, y_test_pred)  # Calculate testing accuracy

print(f"Training accuracy: {training_accuracy}")
print(f"Testing accuracy: {testing_accuracy}")
```

#Task 9
Precision: Measures the proportion of correctly predicted positive instances out of all predicted positives. It indicates how precise the model is when it classifies a sample as positive.

Recall: Measures the proportion of actual positive instances that were correctly identified by the model. It indicates how well the model captures all positive cases.


```python
from sklearn.metrics import confusion_matrix

# Assuming you have trained a DecisionTreeClassifier named 'clf'
y_pred = clf.predict(X_test)  # Predictions on the test set
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n", cm)
#Proportion of False Positives:
false_positive_proportion = cm[0, 1] / (cm[0, 1] + cm[1, 1])
print(f"Proportion of False Positives: {false_positive_proportion}")

#Sensitivity (Recall)
sensitivity = cm[1, 1] / (cm[1, 0] + cm[1, 1])
print(f"Sensitivity (Recall): {sensitivity}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already calculated the confusion matrix 'cm'
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Assuming your DecisionTreeClassifier model is stored in a variable named 'clf'
plt.figure(figsize=(12, 8))  # Adjust figure size for better readability
tree.plot_tree(clf, filled=True, feature_names=X.columns, class_names=['Class 0', 'Class 1'])
plt.show()
```

```python
# Task 12 : Identify and Rank Important Features from the Decision Tree

# Get feature importance scores from the Decision Tree model
feature_importances = clf.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Display ranked features
print(feature_importance_df)
```

```python
#Task
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
# Predict and evaluate
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))

# Plot confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Oranges',
            xticklabels=["Non-Diabetic", "Diabetic"], yticklabels=["Non-Diabetic", "Diabetic"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("KNN Confusion Matrix")
plt.show()
```

```python
# Task 14: Load Data and Identify Two Geographic Features
df = pd.read_csv('/content/drive/MyDrive/pm25_2016_2020.csv')
print(df.columns)
```

```python
  # Task 15 : Extract Two Geographic Columns into a New DataFrame

X = df[['latitude_x', 'longitude_x']]

# Display first few rows
print(X.head())
```

```python
# Task 17 : Modify the Code to Use 5 Clusters and Count Observations
# Apply KMeans with 5 clusters
from sklearn.cluster import KMeans # Import the KMeans class
kmeans = KMeans(n_clusters=5, random_state=1) # Initialize KMeans
X["Cluster"] = kmeans.fit_predict(X)
X["Cluster"] = X["Cluster"].astype("category")

# Count number of observations per cluster
cluster_counts = X["Cluster"].value_counts()
print(cluster_counts)
```

```python
# Task 18 : Fix and Run the Cluster Visualization Code
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(12,8))

# Plot scatter plot with clusters
sns.scatterplot(x=X["longitude_x"], y=X["latitude_x"], hue=X["Cluster"], palette="viridis")

# Add labels and title
plt.xlabel("longitude_x")
plt.ylabel("latitude_x")
plt.title("Geographic Clustering of Weather Monitoring Stations")

# Show plot
plt.show()
```
